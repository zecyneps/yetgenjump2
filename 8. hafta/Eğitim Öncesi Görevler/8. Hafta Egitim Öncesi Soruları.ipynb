# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
file = open("urunler.txt", "w" , encoding="utf-8")
file.close()
def ekle():
    while True:
        urun = (input("Ürün adını ve fiyatını giriniz. Çıkmak için q tuşlayınız:"))
        if urun == "q":
            break
        else:
            global file
            with open("urunler.txt", "r+", encoding="utf-8") as file:
                file.write(urun)
                file.seek(0)
                print(file.readlines())
        
    

ekle()

````

# Problem 2
1.problemde oluşturulan text dosyası içerisinde kullanıcıdan alacağımız eski kelime isimli değişkene atadığımız 
ve değiştirilmesini istediğimiz kelimemizi yerini alıcak olan yeni kelime isimli değişkene atayın. Belirlenen kelimeleri text dosyasında değiştiren fonksiyonu yazınız.

````python

with open("urunler.txt", "r+", encoding="utf-8") as file:
    icerik = file.read()
    def kontrol():
        global icerik
        eski_kelime = input("Değiştirmek istediğiniz kelimeyi giriniz.")
        if eski_kelime in icerik:
            print(f"eski içerik: {icerik}")
            yeni_kelime = input("Yeni kelimeyi giriniz.")
            yeni_icerik = icerik.replace(eski_kelime, yeni_kelime)
            print(f"yeni içerik: {yeni_icerik}")
            with open("urunler.txt", 'w') as file:
                file.write(yeni_icerik)
                print(f"Ürünler.txt dosyasında {eski_kelime} kelimesi {yeni_kelime} ile değiştirildi.")
        else:
            print("Girdiğiniz kelime dosyanın içinde bulunmamaktadır. Tekrar deneyiniz.")
            kontrol()
kontrol()




````

# Problem 3
Vehicle sınıfının tüm değişkenlerini ve yöntemlerini devralacak bir alt sınıf olarak Bus sınıfı oluşturunuz.



```
Output
> Name: School Volvo Speed: 180 Mileage: 12
```


````python
class Bus():
    def __init__(self, name, speed, mileage):
        self.name = name
        self.speed = speed
        self.mileage = mileage
oto = Bus("School Volvo", "180", "12")
print(f"Name: {oto.name} Speed: {oto.speed} Mileage: {oto.mileage}.")
````
