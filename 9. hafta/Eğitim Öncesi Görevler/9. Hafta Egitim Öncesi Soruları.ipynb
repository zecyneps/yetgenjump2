# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicles():
    def __init__(self, type, brand, usage):
        self.type = type
        self.brand = brand
        self.usage = usage
    def intro(self):
        print(self.type, self.brand, self.usage)

class Bus(Vehicles):
    def __init__(self, type, brand, usage, capacity):
        super().__init__(type, brand, usage)
        self.capacity = capacity
    def intro(self):
        print(self.type, self.brand, self.usage, self.capacity)
    def seating_capacity(self):
        print(f"The seating capacity of this {self.type} is {self.capacity} passengers")

b1 = Bus("Bus", "Mercedes", "School", "50")
b1.intro()
b1.seating_capacity()
    
````
